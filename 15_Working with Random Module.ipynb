```python
# If we wnt to generate random value then we can use random module.

# Functions Of Random Module:

# 1.randint() 2.random() 3.randrange() 4.shuffle() 5.uniform() 6.sample() 7.choice()
```

```python
# randint():
# We will pass start value and stop value based on this, it will generate random numbers.

import random
val= random.randint(1,10)
print(val)
```

```python
# random():It will generate a random float number between 0 and 1.
val=random.random()
print(val)
```

```python
#randrange():Exactly same as randint but herew stop valut included.
val=random.randrange(1,10) # Here 10 is not included.
print(val)
```

```python
# Without shuffle:
l=[10,20,30,40,50,60]
print(l)
```

```python
# With shuffle:
l=[10,20,30,40,50,60]
random.shuffle(l)
print(l)
```

```python
# Uniform():It will generate float value in between that range.
val=random.uniform(10,20)
print(val)
```

```python
print(help(random.uniform))
```

```python
# Sample: It will generate sample from data.
li=['surendra','priyanka','rahul','zini','jack','scoot','dev','anjali']
val=random.sample(li,3)
print(val)
```

```python
print(help(random.sample))
```

```python
data=random.sample(range(10000000),60)
print(data)
```

# Number Guessing Game:

```python
import random
rn=random.randint(1,10)
while True:
    choice=int(input("Guess a number in between 1 to 10: "))
    if choice==rn:
        print("Cong!!! you guessed correct number")
        break
    elif choice>rn:
        print("Sry!!! your guess number is greater than computer generated number")
    else:
        print("Sry!!! your guess number is lesser than computer generated number")
        
        
```
